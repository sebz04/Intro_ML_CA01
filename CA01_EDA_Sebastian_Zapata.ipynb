```python
#Finding missing values

pd.set_option('display.max_rows', None)
df.isnull().sum().sort_values(ascending=False)
```

```python
#MasVnrArea --> most missing values with 8
df = df.dropna(subset=['MasVnrArea'])
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(all_cols):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(col)

# remove extra subplot if needed
fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

```python
#AI formulated function

def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    return outliers, lower_bound, upper_bound
```

```python
outlier_summary = {}

for col in cols:
    outliers, lower, upper = detect_outliers_iqr(df, col)
    outlier_summary[col] = {
        'OutlierCount': outliers.shape[0],
        'LowerBound': lower,
        'UpperBound': upper
    }

outlier_df = pd.DataFrame(outlier_summary).T
outlier_df
```

```python
#Outliers were detected using lowerbounds and upperbounds
#OverallQuality and Year Built rarely had outliers
#Total Basement Squarefootage and Lot Area had most amount of outliers
```
